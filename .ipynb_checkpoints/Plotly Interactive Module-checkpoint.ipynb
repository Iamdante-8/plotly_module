```python
import plotly.graph_objects as go
df_stocks=px.data.stocks()
px.line(df_stocks,x='date',y='GOOG',labels={'x':'Date','y':'Price'})
px.line(df_stocks,x='date',y=['GOOG','AAPL'],labels={'x':'Date','y':'Price'},title='GOOGLE VS. APPLE')

fig=go.Figure()
fig.add_trace(go.catter(x=df_stocks.date,y=df_stocks.AAPL,
                         mode='line',name='Apple'))
fig.add_trace(go.scatter(x=df_stocks.date,y=df_stocks.AMZN,
                         mode='line+markers',name='Amazon'))
fig.add_trace(go.scatter(x=df_stocks.date,y=df_stocks.GOOG,
                         mode='line+markers',name='Google',line=dict(color='firebricks',
                                                                     width=2,dash='dashdot')))
```
